```python
%%writefile shuffle_demo.py
import numpy as np
ran=np.random.default_rng()
print(ran.integers(6))
print(ran.uniform(low=10,high=90,size=10))
laptops=np.array(["HP","Lenovo","Dell","MacAir","Toshiba"])
ran.shuffle(laptops,axis=0)
print(laptops)
print(ran.choice(laptops))
```
